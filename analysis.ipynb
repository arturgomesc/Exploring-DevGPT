```python
import os
import json
from glob import glob
import pandas as pd
```

```python
def get_unique_authors(data_folder):
    json_file_patterns = [
        '*_hn_sharings.json',
        '*_pr_sharings.json',
        '*_issue_sharings.json',
        '*_discussion_sharings.json',
        '*_commit_sharings.json',
        '*_discussion_sharings.json'
    ]
    
    unique_authors = set()

    for root, dirs, files in os.walk(data_folder):
        for pattern in json_file_patterns:
            for json_file in glob(os.path.join(root, pattern)):
                try:
                    df = pd.read_json(json_file)
                    if 'Sources' in df.columns:
                        sources_df = pd.json_normalize(df['Sources'])
                        if 'Author' in sources_df.columns:
                            authors = sources_df['Author'].dropna().unique()
                            unique_authors.update(authors)
                        else:
                            print(f"'Author' column not found in {json_file}")
                    else:
                        print(f"'Sources' key not found in {json_file}")
                except Exception as e:
                    print(f"Error processing file {json_file}: {e}")
                    continue


    return unique_authors

if __name__ == '__main__':
    data_folder = 'snapshots'
    authors = get_unique_authors(data_folder)
    print(f'Total unique users who interacted: {len(authors)}')
    print(authors)

```
